```python
#importing neccessary libraries
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None

import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.style.use('ggplot')
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#import the data
df = pd.read_csv(r'C:\Users\abdul\Downloads\DA Prjs\moviesAnalysis\movies.csv')
```

```python
#looking at the data
df.head()
```

```python
#looking for missing data
for col in df.columns:
    prcnt_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(prcnt_missing*100)))
```

```python
#data types of the columns
df.dtypes
```

```python
df['budget'] = df['budget'].fillna(0)
df['gross'] = df['gross'].fillna(0)

df.head(10)
```

```python
#changing data types
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df.head(10)
```

```python
#budget VS gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget VS Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df, line_kws={'color': 'black'})
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

### We can see that there is high correlation between budget and gross

```python
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_num = df

for col in df_num.columns:
    if(df_num[col].dtype == 'object'):
        df_num[col] = df_num[col].astype('category')
        df_num[col] = df_num[col].cat.codes
df_num.head()
```

```python
correlation_matrix = df_num.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
corr_pairs = df_num.corr().unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```

### We can see that votes are also highly correlated to gross

### The conclusion is that, the higher the votes and the budget, the higher the the gross earnings of the movie.
